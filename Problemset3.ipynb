```python
#Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'
dataframe_nw = nw.bedrs.astype(str).str.cat(nw.bathrs.astype(str)).str.cat(nw.price_sqr_meter.astype(str))
dataframe_newcols = ['bigcolumn']
dataframe_nw
```

```python
#it is going only until index 99
dataframe_nw = dataframe_nw.reset_index(drop=True)
dataframe_nw
```

```python
#Reindex the DataFrame so it goes from 0 to 299
dataframe_nw.reset_index()
dataframe_nw.reindex(index=range(0,299))
```

```python
import pandas as pd
import random
import matplotlib.pyplot as plt
da = pd.read_csv("roman-emperor-reigns.csv")
da.head()
ld= da.where(da.Cause_of_Death=="Assassinated").Cause_of_Death.count()
rg= da.Cause_of_Death.count()- ld
label= ["Other Causes of Deaths","Assassinated"]
plt.pie([rg,ld],labels = label,autopct='%.2f%%')
plt.title("Roman Emperors")
plt.show()
```
